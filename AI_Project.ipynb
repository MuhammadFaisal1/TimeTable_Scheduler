```python
def cross_over(p1,p2):
    for i in range(5):
        for j in  range(int(Total_rooms)):
            if np.shape(p1[i][j])[0]>0 and np.shape(p2[i][j])[0]>0:
                if np.shape(p1[i][j])[0]> np.shape(p2[i][j])[0]:
                    split = np.random.randint(np.shape(p2[i][j])[0])
                    for l in range(split,np.shape(p2[i][j])[0]):
                        temp = p1[i][j][l]
                        p1[i][j][l] = p2[i][j][l] 
                        p2[i][j][l] = temp
                if np.shape(p1[i][j])[0]< np.shape(p2[i][j])[0]:
                    split = np.random.randint(np.shape(p1[i][j])[0]) 
                    for l in range(split,np.shape(p1[i][j])[0]):
                        temp = p1[i][j][l]
                        p1[i][j][l] = p2[i][j][l] 
                        p2[i][j][l] = temp
                if np.shape(p1[i][j])[0] == np.shape(p2[i][j])[0]:
                    split = np.random.randint(np.shape(p2[i][j])[0])       
                    for l in range(split,np.shape(p1[i][j])[0]):
                        temp = p1[i][j][l]
                        p1[i][j][l] = p2[i][j][l] 
                        p2[i][j][l] = temp
    return p1,p2            
```

```python
#A simple and smoothe algorithm for mutation
def mutation(parent1,parent2):
    for index in range(2):
        if index==0:
            a = parent1
        else:
            a = parent2
        for j in range(5):
            for k in  range(int(Total_rooms)):
                m = np.shape(a[j][k])[0]
                for l in range(m):
                    c = np.random.randint(m)
                    r = np.random.randint(Instructors)
                    if r>1:
                        a[j][k][c].Instructors = 'T'+str(r)
                    else:
                        while(r<1):
                            r = np.random.randint(Instructors)
                        a[j][k][c].Instructors = 'T'+str(r)
        if index==0:
            parent1 = a
        else:
            parent2 = a
    return parent1,parent2        
```

```python
def Genetic_Algorithm(population,generations):
    #Reproduce and regenerate whole population from this one small part
    populized_data = initialize_population(population)
    #We generate the whole thing using population as many times as generators run off.
    for gen in range(generations):
        #Function call to determine fitness for the initialized population
        fitness_val=Determine_fitness(populized_data)
        #take a humble start every time: Use argmin
        minimum_fitness = np.argmin(fitness_val)
        if len(populized_data)>1:
            for index in range(len(populized_data)/2):
                if (index*2)+1<len(populized_data)-1:
                    cp1,cp2 = cross_over(populized_data[index*2],populized_data[(index*2)+1])
                    mp1,mp2 = mutation(cp1,cp2)
                    del populized_data[index*2]
                    del populized_data[(index*2)+1]
                    populized_data.append(mp1)
                    populized_data.append(mp2)
    fitness_val=Determine_fitness(populized_data)  
    fit = np.argmin(fitness_val)
    #print fit
    return populized_data[np.argmax(fitness_val)]            
```
